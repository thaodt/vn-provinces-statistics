## Read csv file to find 10 largest Vietnam provinces by 2015. Then draw the bar graph of the area of those data.
### Data is provided via  https://www.gso.gov.vn/

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('dientich.csv', 
                 sep=';', 
                 index_col=0, 
                 usecols=[0,5], 
                 skiprows=[0, 3],
                 skipfooter=4,
                 engine='python',
                 dtype={'2015':np.float64})
```

```python
df
```

```python
data = df.nlargest(10, columns='2015')
```

```python
data
```

```python
# drawing
data.plot.barh(title='10 biggest VN provinces area in 2015')
```
